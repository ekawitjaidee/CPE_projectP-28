```python
import pandas as pd
import glob as glb
import numpy as np
from collections import Counter

from Indicator import Indicator
from GA import GA

def norm(df,col_name):# this will get columnname + _n
    #Normalize data each colunm in to range -1 to 1
    df[col_name+'_n'] = 2*(df[col_name]-df[col_name].min())/(df[col_name].max()-df[col_name].min())-1    
    return df
```

```python
datalist = [f for f in glb.glob('GT/*.csv')]#Read data name
len(datalist)
```

## Find the most profitable parameter of MACD Indicator
MACD Indicator have 3 parameter then we use Genetic Algorithm to Random Number For MACD parameter

```python
paramlist = []
data = pd.DataFrame([])
for name in datalist:
    print('Stock: ',name,' In Progress ...',)
    df = pd.read_csv(name)
    pop = GA().first_generation(200,df)
    pop[-1] = [[12,26,9],0]
    pop = GA().fit(pop = pop,Maxgen = 50,Elite = 10,df = df)
    paramlist.append([name[3:-4],pop[0]])
    print('Done!!')
```

```python
k = []
for it in paramlist:
    k.append(it[1][0])
```
